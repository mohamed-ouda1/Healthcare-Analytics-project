# Data Cleaning


To ensure the integrity of our analysis, we will:
1. **Handle Outliers**: Remove negative billing amounts.
2. **Normalize Text**: Standardize names and categorical data.
3. **Consistency Check**: Ensure stay duration is positive.

```python
# 1. Filter out negative or zero billing amounts
initial_count = len(df)
df = df[df['Billing Amount'] > 0]
print(f'Removed {initial_count - len(df)} rows with invalid billing amounts.')

# 2. Normalize Name casing
df['Name'] = df['Name'].str.title()

# 3. Verify length of stay is positive
df = df[df['length_of_stay'] >= 0]
print(f'Final dataset contains {len(df)} records.')
```

# check for null values

```python
df.isnull().sum()
```

```python
df.columns
```

### Remove duplication

```python
before = df.shape[0]
df = df.drop_duplicates()
after = df.shape[0]

before, after
```

```python
bins = [0, 18, 30, 45, 60, 100]

labels = ['0-18', '19-30', '31-45', '46-60', '60+']

df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)
```

```python
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
```

```python
df['length_of_stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

# Save the new file

```python
output_path = "clean_healthcare__dataset.csv"
df.to_csv(output_path, index=False)
```

# Exploratory Data Analysis (EDA)

### 📊 Demographic Analysis

```python
plt.figure(figsize=(15, 5))

# Gender Distribution
plt.subplot(1, 2, 1)
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Patient Gender Distribution')

# Age Group Distribution
plt.subplot(1, 2, 2)
sns.countplot(x='age_group', data=df, palette='magma')
plt.title('Age Group Distribution')

plt.tight_layout()
plt.show()
```

###  Cost Analysis by Admission Type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Admission Type', y='Billing Amount', data=df, palette='Set2')
plt.title('Billing Amount Distribution by Admission Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###  Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
performance = df.groupby('Hospital').agg({
    'Billing Amount': 'mean',
    'length_of_stay': 'mean',
    'Name': 'count'
}).rename(columns={'Name': 'Patients'})
```

```python
performance['Score'] = (
    performance['Patients'] * 0.3 +
    performance['Billing Amount'] * 0.4 +
    performance['length_of_stay'] * 0.3
)

performance.sort_values('Score', ascending=False).head(10)
```

```python
hospital_cost = df.groupby('Hospital')['Billing Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=hospital_cost.values, y=hospital_cost.index, palette='Blues_r')
plt.title('Top 10 Hospitals by Total Billing')
plt.xlabel('Total Billing Amount')
plt.ylabel('Hospital Name')
plt.show()
```

## 🎯 4. Final Conclusion & Business Recommendations

### 📌 Summary of Findings
- **Cost Variability**: Admission type is the primary driver of cost variability, with emergency cases requiring more resources and higher billing.
- **Demographic Trends**: The patient population is well-distributed across genders, but certain age groups show higher frequency of specific medical conditions.
- **Operational Efficiency**: Length of stay is highly correlated with cost, suggesting that optimizing discharge processes could reduce financial burden on patients.

### 🚀 Recommendations for Stakeholders
1. **Emergency Department Optimization**: Review resource allocation in emergency units to manage high-cost cases more efficiently.
2. **Insurance Partnership**: Collaborate with insurance providers to create better coverage plans for the most frequent medical conditions (e.g., Cancer, Diabetes).
3. **Predictive Stay Modeling**: Implement models to predict length of stay upon admission to improve bed management and cost estimation.
